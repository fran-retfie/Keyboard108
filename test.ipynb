```python
import pcbnew as pcb
```

```python
file = pcb.LoadBoard("keyboard.kicad_pcb")
```

```python
board = file.GetBoard()
```

```python
for footprint in board.GetFootprints():
    if(footprint.GetValue() == "SK6812MINI"):
        print(footprint.GetPosition())
        p = footprint.GetPosition()
        p.x = p.x + 10000000
        footprint.SetPosition(p)
        print(footprint.GetPosition())
```

```python
pcb.SaveBoard("k.kicad_pcb", board)
```
